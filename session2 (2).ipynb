```python
iso_codes = {}

for index, name in df['Country (according to CN Gov):'].items():
    if name in iso_codes:
        continue
    
    cleaned_name = name.split('(')[0].strip()
    if cleaned_name == 'Thailiand':
        cleaned_name = 'Thailand'
    elif cleaned_name == 'The United Arab Emirates':
        cleaned_name = 'United Arab Emirates'
    
    url = 'https://restcountries.eu/rest/v2/name/' + cleaned_name
    r = requests.get(url)
    if r.status_code == 200:
        data = r.json()
        country_code = data[0]['alpha3Code']
        iso_codes[name] = country_code
    else:
        print('[Error] we cannot get the data, the reason is',r.status_code, r.text)
        print(name, cleaned_name)
```

```python
df['isocode'] = df['Country (according to CN Gov):'].apply(lambda x: iso_codes[x])
```

```python
WorldBank = {'IMF WB\nCustoms and other import duties (% of tax revenue) 2016': 'GC.TAX.IMPT.ZS', 'WB IMF Non-performing loans to total gross loans ratio (%)': 'FB.AST.NPER.ZS', 'UIC Rail freight (mn ton x km travelled)': 'IS.RRS.GOOD.MT.K6', 'ICAO air freight (mn ton x km travelled)': 'IS.RRS.GOOD.MT.K6', 'WB Household final consumption expenditure, PPP bn USD': 'NE.CON.PRVT.PP.CD', 'WB UNESCO R&D Expenditure % of GDP': 'GB.XPD.RSDV.GD.ZS', 'WB UNESCO Researchers p. mn inhabitants': 'SP.POP.SCIE.RD.P6', 'IFDI Net Inflows (% of GDP)': 'BX.KLT.DINV.WD.GD.ZS', 'WHO Current Health Expenditure p.c., PPP USD': 'SH.XPD.CHEX.PP.CD', 'WB Research Expenditure % of GDP': 'GB.XPD.RSDV.GD.ZS', 'WIPO Trademark applications': 'IP.TMK.RSCT'}

for Indicator, CodeURL in WorldBank.items():
    ValueIndicators[Indicator] = {}
    for i, iso_codes in df['isocode'].items():
        url = 'https://api.worldbank.org/v2/en/country/' + iso_codes + '/indicator/' + CodeURL + '?format=json&per_page=20000&source=2&mrv=1'
        r = requests.get(url)
        print (iso_codes)
        if r.status_code == 200:
            data = r.json()
            if len(data) == 2:
                page = data[0]['page']
                if page == 1 :
                    value = data[1][0]['value']
                    ValueIndicators[Indicator][iso_codes] = value
                else:
                    print('erreur')
                    ValueIndicators[Indicator][iso_codes] = 'null'
            else:
                print('erreur')
                ValueIndicators[Indicator][iso_codes] = 'null'
        else:
            ValueIndicators[Indicator][iso_codes] = 'null'
            print('[Error] we cannot get the data, the reason is',r.status_code, r.text)
            print(iso_codes)
            
print(ValueIndicators)
```
